Handling missing values: 
1. Deleting the columns with missing values.
(the problem with this method is that we may lose valuable information on that feature, as we have deleted it completely due to some null values)
2. Delecting the rows with the missing values.(in this case there are possibiliies of getting better accuracy than before.this might be because the columns contains more valuable information than we expected)

3. Imputation method: filling the missing data with the mean or median value if its a numerical variable.
4. filling the missing data with mode if its a categorical value.
5. filling the numerical value with 0 or -999 or some other number that will not occur in the data,this can be done so that the machine can recognise that the data is not real or is different.
6. filling the categorical value with a new type for the missing values.


```python
#deleting the columns with missing values
updated_df1 = df.dropna(axis=1)
updated_df1.info()
```

here the null values containing columns are all deleted as of our data the firstname and lastname of that data are important columns so it is not a best way of handling missing values.

```python
#deleting the rows with the missing values
updated_df2 = df.dropna(axis=0)
updated_df2.info()
```

in this we have all the columns and the rows which has the null values have been deleted.

```python
updated_df1  = df
#as we have categorical data Firstname and lastname we will fill the missing values with unknown
updated_df1['Firstname'] = updated_df1['Firstname'].fillna("unknown")
updated_df1['Lastname'] = updated_df1['Lastname'].fillna("unknown")
updated_df1.info()
```

```python
#handling missing values of age(with mean)
#filling the missing values with mean as the data is numerical and the mean is not affected by outliers.
df['Age'].mean()
updated_df1['Age'] = updated_df1['Age'].fillna(df['Age'].mean())
```
